Import the pandas_tasks.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

data2 = pd.read_csv("pandas_tasks.csv")
```

#### Question 1. Make a scatter plot of the Longitude and Latitude columns

```python
sns.relplot(x="Longitude", y="Latitude", height= 8, aspect=1, data=data2)
```

#### Question 2. Make a scatter plot of the Longitude and Latitude columns. This time, plot with 'hue' = Q2 and compare it over Q3

```python
sns.relplot(x="Longitude", y="Latitude", height= 8, aspect=1, hue="Q2", col='Q3',data=data2)
```

#### Question 3. Make a boxplot of Longiture and Q2

```python
sns.boxplot(x='Longitude', y='Q2', data=data2)
```
