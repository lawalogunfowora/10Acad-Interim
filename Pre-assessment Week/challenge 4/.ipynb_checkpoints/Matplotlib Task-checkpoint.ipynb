## Boxplot

Using the latitude column

```python
plt.boxplot(data['Latitude'], widths=0.1, patch_artist=True, boxprops=dict(facecolor="red"))
plt.title('Boxplot of the Approximate latitude of the dataset')
plt.ylim(-15,0)
plt.ylabel("Latitude")
plt.show()
```

## Bar Chart

Using Q3, Q1, Q16 & Q17

```python
plt.bar(data['Q3'], data['Q1'], color='green', width=0.4)
plt.ylim(0,110)
plt.title("A Bar plot of marital status and age")
plt.xlabel("Marital Status")
plt.ylabel("Ages")
positions = (1, 2, 3, 4)
labels = ("Married", "Divorced", "Widowed", "Single")
plt.xticks(positions, labels)
plt.show()
```

```python
goods_services = data['Q16'][:10]
bills = data['Q17'][:10]
ind = np.arange(10) # creates a numpy array from 0 to 3. it will be used on the x-axis
width = 0.3 # will be used to specify the width of the bar
# plt.xticks(ind + width / 2, ('Not Applicable', 'Never', 'Daily', 'Weekly', 'Monthly', 'Less often than monthly'), rotation=50)
plt.bar(ind, goods_services, width, label="Goods/Services")
plt.bar(ind + width, bills, width, label="Bills")
plt.title("A Bar Plot comparing the the use of mobile for paying bills and purchasing goods/services")
plt.xlabel("Value")
plt.ylabel("Count")
plt.ylim(-2, 6)
plt.legend()
plt.show()
```

## Histogram
Using Ages

```python
plt.hist(data['Q1'], bins=30, histtype='step', color='red')
plt.title("A Histogram showing the distribution of the ages in the dataset")
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.ylim(0, 900)
plt.xlim(10, 110)
plt.show;
```

## Scatter Plot

```python
plt.scatter(data['Latitude'], data['Longitude'], color='red')
plt.title("A scatter plot showing the relationship between Latitude and Longitude")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.ylim(28, 43)
plt.show()
```
