# Solution to Question 2

#### The hypothesis statement can be written as: 

**H0 : There is no significant difference between the means of children born from smoking versus non-smoking mothers.**

**HA : There is a significant difference between means.**

```python
import statsmodels.stats.weightstats as st

nc_weight_smoker = nc[nc['habit'] == 'smoker']['weight']
nc_weight_nonsmoker = nc[nc['habit'] == 'nonsmoker']['weight']

d_stat = st.DescrStatsW(nc_weight_smoker)
d_stat_2 = st.DescrStatsW(nc_weight_nonsmoker)
mean_compare = st.CompareMeans(d_stat_2,d_stat)

# calculate number of observations, mean and standard deviation for each group
n_smoker = d_stat.nobs
n_nonsmoker = d_stat_2.nobs
mean_smoker = d_stat.mean
mean_nonsmoker = d_stat_2.mean
sd_smoker = d_stat.std
sd_nonsmoker = d_stat_2.std
print(f'n_smoker = {n_smoker}')
print(f'mean_smoker = {mean_smoker}')
print(f'sd_smoker = {sd_smoker}')
print()
print(f'n_nonsmoker = {n_nonsmoker}')
print(f'mean_nonsmoker = {mean_nonsmoker}')
print(f'sd_nonsmoker = {sd_nonsmoker}')
print()
# conduct hypothesis test
ht = mean_compare.ztest_ind(alternative = 'two-sided', usevar = 'unequal', value = 0)

# calculate and print the standard error, the Z-score, and p-value for the hypothesis test
se = mean_compare.std_meandiff_separatevar
testZ = ht[0]
p_value = ht[1]
print(f'my Standard error is = {se}')
print(f'my Test statistic is: Z = {testZ}')
print(f'my p-value is = {p_value}')

# I then, set criteria to reject or fail to reject null hypothesis
if(p_value) < 0.05:
    print('The verdict is to reject null hypothesis')
else:
    print('The verdict is fail to reject null hypothesis')
```

**As seen in the result above,  the P-value is less than 0.05, Thus, we reject the null hypothesis.**

# Solution to Question 3

```python
import statsmodels.api as sm

nc_younger_mom = nc[nc['mature'] == 'younger mom']
nc_mature_mom = nc[nc['mature'] == 'mature mom']
n1 = len(nc_younger_mom)
mu1 = nc_younger_mom['gained'].mean()
sd1 = nc_younger_mom['gained'].std()

print(f'n_younger_mom = {n1}')
print(f'mean_younger_mom = {mu1}')
print(f'sd_younger_mom = {sd1}')
print()

n2 = len(nc_mature_mom)
mu2 = nc_mature_mom['gained'].mean()
sd2 = nc_mature_mom['gained'].std()

print(f'n_mature_mom = {n2}')
print(f'mean_mature_mom = {mu2}')
print(f'sd_mature_mom = {sd2}')
print()

# conducting my hypothesis test
ht = sm.stats.ztest(nc_younger_mom['gained'].dropna(), nc_mature_mom['gained'].dropna(),alternative='two-sided')
testZ = ht[0]
p_value = ht[1]
print(f'my Test statistic is: Z = {testZ}')
print(f'my p-value is = {p_value}')


# I then, set criteria to reject or fail to reject null hypothesis
if(p_value) < 0.05:
    print('The verdict is: reject null hypothesis')
else:
    print('The verdict is: fail to reject null hypothesis')
```
